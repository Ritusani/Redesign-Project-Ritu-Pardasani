# Data cleaning and wrangling for the datasets used for the project:

```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('electionsdata.csv')
```

```python
df.head()
```

```python
df  = df[['State','Margin','Clinton','Trump']]
```

```python
df.head()
```

```python
## checking the diffrence of votes between Trump and clinton to check who won. 
win = []
for i in range(len(df)) : 
    margin = int(df.Clinton[i].replace(",","")) - int(df.Trump[i].replace(",",""))
    win.append('Clinton' if (margin > 0) else 'Trump' )
```

```python
df['win']= win
```

```python
df.head()
```

```python
df.to_csv('elections_final.csv')
```
